```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('students_data.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={
    'student_id':'ID',
    'name':'Name',
    'age':'Age',
    'gender':'Gender',
    'grade':'Grade',
    'math_score':'Math_Score',
    'english_score':'English_Score',
    'science_score':'Science_Score',
    'enrolled_date':'Enrolled_Date'
})
df
```

```python
df['Gender'] = df['Gender'].replace({"male":'M',"Male":'M',"MALE":'M',"female":'F',"Female":'F',"FEMALE":'F',})
df
```

```python
df[['First_Name','Last_Name']] = df['Name'].str.split(r'\s+',n=1,expand=True)
df
```

```python
df.drop(columns=['Name'],inplace=True)
df
```

```python
df = df[['ID','First_Name','Last_Name'] + [c for c in df.columns if c not in ['ID','First_Name','Last_Name','Name']]]
df
```

```python
df['First_Name'] = df['First_Name'].str.capitalize()
df
```

```python
df['Last_Name'] = df['Last_Name'].str.capitalize()
df
```

```python
df['Last_Name'] = df['Last_Name'].replace({"O’reilly":'O-reilly'})
df
```

```python
df['English_Score'] = df['English_Score'].replace({"missing":'77',None:'77'})
df
```

```python
df['Enrolled_Date'] = df['Enrolled_Date'].replace({"2022-06-10":"10-06-2022","06/12/2022":"06-12-2022","2022/06/11":"11-06-2022","2022-06-11":"11-06-2022"})
df
```

```python
df = df.rename(columns={'Grade':'Class'})
df
```

```python
df['Class'] = (df['Class'].astype(str).str.replace(r"(st|nd|rd|th)","",regex=True).astype(int))
df
```

```python
df['Class'] = df['Class'].replace({10:"X",11:"XI",12:"XII"})
df
```

```python
df['Age'] = df['Age'].fillna(16).astype(int)
df
```

```python
df = df.rename(columns={'remarks':'Grade'})
df
```

```python
df['Grade']= df['Grade'].replace({"excellent":"A+","GOOD":"A","Good":"A","needs improvement":"C","Average":"B","average":"B","poor":'D',"good student":"A"})
df
```

```python
df['Math_Score'] = df['Math_Score'].fillna(74).astype(int)
df
```

```python
df[['English_Score','Science_Score']] = df[['Science_Score','English_Score']]
df
```

```python
df = df.rename(columns={'English_Score':'Science_Score','Science_Score':'English_Score'})
df
```

```python
df.drop_duplicates()
```

```python
df.to_csv("students_cleaned_data.csv",index=False, encoding="utf-8-sig")
```
